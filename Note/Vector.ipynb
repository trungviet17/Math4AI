## 3. Transpose operation 

- Phép chuyển vị là phép đổi chiều của vector 
$$ [4 \ 3 \ 0] ^T = \begin{bmatrix} 4 \\ 3 \\ 0 \end{bmatrix} $$
- Phép chuyển vị 2 lần là chính giá trị vector ban đầu $$ v^{TT} = v $$

```python
v1 = np.array([[4, 3, 0]])
v2 = v1.T
print(f'Original vector: {v1.shape}')
print(f'Transpose vector: {v2.shape}')
```

## 4. Operations on Vector


### Vector Addition
- Cộng, trừ các vector là cộng, trừ tương ứng những phần tử của chúng lại với nhau
- Hai vector chỉ cộng được với nhau khi chúng có cùng chiều, cùng hướng

```python
v = np.array([[4, 5, 6]])
w = np.array([[10, 20, 100]])
print(f'v plus w = {v + w}')
```

```python
# Ngoại lệ boardcasting trong Python
print(f'v plus w.transpose = {v + w.T}')
```

```python
def create_vector(mag: int, angle_degrees: int) : 
    angle_radians = np.radians(angle_degrees)
    x_comp = mag * np.cos(angle_radians)
    y_comp = mag * np.sin(angle_radians)
    return x_comp, y_comp


# Create a figure and axis
fig, ax = plt.subplots()

x_1, y_1 = create_vector(3, 30)
x_2, y_2 = create_vector(4, 60)

# Plot the vector
ax.quiver(0, 0, x_1, y_1, angles='xy', scale_units='xy', scale=1, color='blue')
ax.quiver(x_1, y_1, x_2, y_2, angles = 'xy', scale_units = 'xy', scale = 1, color = 'green')
ax.quiver(0, 0, x_2 + x_1, y_2 + y_1, angles = 'xy', scale_units = 'xy', scale = 1, color = 'red')
# Set the aspect ratio of the plot to be equal
ax.set_aspect('equal')

# Set limits
ax.set_xlim([0,6])
ax.set_ylim([0, 6])

# Add gridlines
ax.grid(True)

# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title("Adding two vector")

# Show plot
plt.show()
```

### Vector-Scalar Multiplication 
- Nhân lần lượt giá trị phần tử của vector với giá trị vô hướng 

```python
print(f'Scalar Multiplication of vector {v} with {scalar} is {v * scalar}')
```

- Phép nhân với số âm có thể làm vector bị đổi hướng, tuy nhiên vector khi xoay này không được xem là đổi hướng nếu người ta xét định nghĩa hướng là phương của vector kéo dài về 2 phía

```python
fig, ax = plt.subplots(nrows= 1, ncols= 5)

lst = [1, 2, 1/3, 0, -2]
fig.set_figheight(7)
fig.set_figwidth(20)

for i in range(len(lst)) : 
    x_1, y_1 = create_vector(2 , 45)
    x_2, y_2 = create_vector(2 * lst[i], 45)
    ax[i].quiver(0, 0, x_1, y_1, angles='xy', scale_units='xy', scale=1, color='blue')
    ax[i].quiver(0, 0, x_2, y_2, angles= 'xy', scale_units = 'xy', scale = 1, color = 'grey')
    ax[i].set_aspect('equal')
    ax[i].set_xlim([-3, 3])
    ax[i].set_ylim([-3,  3])
    ax[i].grid(True)
    ax[i].set_title(f'a = {lst[i]}')
    
plt.show()
```

- Kết hợp giữa nhân vô hướng vector với cộng 2 vector dẫn tới vector trung bình được tính giống như trung bình giữa các số

```python
print(f'Vector Average of {v} and {w} is {(v + w) * .5}')
```

## 5. Vector Magnitude and Unit Vectors 
- Độ lớn của một vector được gọi là chiều dài của nó hoặc khoảng cách từ điểm đầu đến cuối của vector. 
- Nó được tính thông qua phép tính Euclidean : tính thông qua căn bậc hai của tổng bình phương khoảng cách các chiều $$  \| v \| =  \sqrt {\sum^n_{i = 1} v_i^2} $$

```python
v = np.array([4, 10, 6, 8, 9, 2])
v_dim = len(v)
print(f'The number of dimension of {v} is {v_dim}')
v_mag = np.linalg.norm(v)
print(f'Magnitude: {v_mag}')
```

- Vector đơn vị là vector có chiều dài là 1. Vector đơn vị có thể được tạo ra bởi vector chiều dài thông qua công thức $$ \hat{v} = \frac {1} {\| v \|}  v $$

```python
print(f'Unit vector of {v} is {1 / v_mag * v}')
```

## 6. Vector Dot product
- Dot product (inner product) là một trong những phép toán quan trọng trong đại số tuyến tính. 
- Kết quả của phép tính là một số cung cấp thông tin về mối quan hệ giữa 2 vector. 
- Để tính dot product, ta nhân tương ứng các phần tử của 2 vector và cộng chúng lại với nhau. $$ \delta = \sum^n_{i = 1} a_ib_i $$

```python
# hàm dot implement matrix multiplication
v = np.array([1, 4, 1, 7, 9])
w = np.array([5, 7, 9, 9, 10])
print(f'The dot production of {v} and {w} is {np.dot(v, w)}')
```

```python
print(f'The dot production of 2 multiplying {v} and {w} is {np.dot(2 * v, w)}')
```

- Dot product có tính kết hợp $ a^T (b + c) = a^Tb + a^Tc $

```python
p = np.array([9, 10, 3, 5, 7])
print(f'v * (w + p) :{np.dot(v, w + p)}')
print(f'v * w + v * p: {np.dot(v, w) + np.dot(v, p)}')
```

- Về hình học, dot product được biểu diễn là tích chiều dài của 2 vector nhân với cos của góc giữa 2 vector đó . $$ \alpha =  cos(\theta_{v, w})\|v\|\|w\|  $$
- Hai công thức đưa ra về mặt đại số và hình học là tương đương với nhau

```python
fig, ax = plt.subplots(nrows = 1, ncols= 5)
fig.set_figheight(5)
fig.set_figwidth(20)

angle = [30, 120, 90, 0, 180]

for i in range(len(angle)) : 
    x_1 , y_1 = create_vector(2, 0)
    x_2, y_2 = create_vector(2, angle[i])
    ax[i].quiver(0, 0, x_1, y_1, angles='xy', scale_units='xy', scale=1, color='blue')
    ax[i].quiver(0, 0, x_2, y_2, angles = 'xy', scale_units='xy', scale=1, color='green')
    ax[i].set_aspect(True)
    ax[i].set_xlim([-3, 3])
    ax[i].set_ylim([-3,  3])
    ax[i].grid(True)
    ax[i].set_title(f'alpha = {angle[i]} \n dot product = {round(np.cos(np.radians(angle[i])) * 2 * 2)}')
plt.show()
```

### Outer Product
- Là cách để tạo ra matrix từ vector cột và vector hàng. Mỗi hàng nhân với mỗi phần tử cột tương ứng tạo thành matrix mới.

## 8. Orthogonal  Vector Decomposition

- Decompose (Phân tích) một vector có nghĩa là chia vector đó thành nhiều phần đơn giản hơn, giúp trích xuất thông tin trong matrix và dễ dàng sử dụng với nó. 
- Orthogonal Vector Decomposition chia vector thành 2 phần, một vector trực giao với vector tham chiếu và một vector song song với vector tham chiếu
- Hình thành khái niệm về Orthogonal Vector decomposition, người ta bắt đầu với khái niệm về hình chiếu vuông góc của vector b lên vector a

- Chiếu b lên a được một vector mới là $\beta a$ khi đó vector khoảng cách từ điểm b tới a là $ b - \beta a$. Nhiệm vụ cần tìm $\beta$ để xác định hình chiếu. Sử dụng dot product khi hai vector trên trực giao với nhau : $$ a^T (b - \beta a) = 0 \\ \Leftrightarrow \beta = \frac {a^Tb} {a^Ta} $$
