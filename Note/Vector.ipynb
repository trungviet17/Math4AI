```python
def create_vector(mag: int, angle_degrees: int) : 
    angle_radians = np.radians(angle_degrees)
    x_comp = mag * np.cos(angle_radians)
    y_comp = mag * np.sin(angle_radians)
    return x_comp, y_comp


# Create a figure and axis
fig, ax = plt.subplots()

x_1, y_1 = create_vector(3, 30)
x_2, y_2 = create_vector(4, 60)

# Plot the vector
ax.quiver(0, 0, x_1, y_1, angles='xy', scale_units='xy', scale=1, color='blue')
ax.quiver(x_1, y_1, x_2, y_2, angles = 'xy', scale_units = 'xy', scale = 1, color = 'green')
ax.quiver(0, 0, x_2 + x_1, y_2 + y_1, angles = 'xy', scale_units = 'xy', scale = 1, color = 'red')
# Set the aspect ratio of the plot to be equal
ax.set_aspect('equal')

# Set limits
ax.set_xlim([0,6])
ax.set_ylim([0, 6])

# Add gridlines
ax.grid(True)

# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title("Adding two vector")

# Show plot
plt.show()
```

## 5. Vector Magnitude and Unit Vectors 
- Độ lớn của một vector được gọi là chiều dài của nó hoặc khoảng cách từ điểm đầu đến cuối của vector. 
- Nó được tính thông qua phép tính Euclidean : tính thông qua căn bậc hai của tổng bình phương khoảng cách các chiều $$  \| v \| =  \sqrt {\sum^n_{i = 1} v_i^2} $$

```python
v = np.array([4, 10, 6, 8, 9, 2])
v_dim = len(v)
print(f'The number of dimension of {v} is {v_dim}')
v_mag = np.linalg.norm(v)
print(f'Magnitude: {v_mag}')
```

- Vector đơn vị là vector có chiều dài là 1. Vector đơn vị có thể được tạo ra bởi vector chiều dài thông qua công thức $$ \hat{v} = \frac {1} {\| v \|}  v $$

```python
print(f'Unit vector of {v} is {1 / v_mag * v}')
```

## 6. Vector Dot product
- Dot product (inner product) là một trong những phép toán quan trọng trong đại số tuyến tính. 
- Kết quả của phép tính là một số cung cấp thông tin về mối quan hệ giữa 2 vector. 
- Để tính dot product, ta nhân tương ứng các phần tử của 2 vector và cộng chúng lại với nhau. $$ \delta = \sum^n_{i = 1} a_ib_i $$

```python
# hàm dot implement matrix multiplication
v = np.array([1, 4, 1, 7, 9])
w = np.array([5, 7, 9, 9, 10])
print(f'The dot production of {v} and {w} is {np.dot(v, w)}')
```

```python
print(f'The dot production of 2 multiplying {v} and {w} is {np.dot(2 * v, w)}')
```

- Dot product có tính kết hợp $ a^T (b + c) = a^Tb + a^Tc $

```python
p = np.array([9, 10, 3, 5, 7])
print(f'v * (w + p) :{np.dot(v, w + p)}')
print(f'v * w + v * p: {np.dot(v, w) + np.dot(v, p)}')
```

- Về hình học, dot product được biểu diễn là tích chiều dài của 2 vector nhân với cos của góc giữa 2 vector đó . $$ \alpha =  cos(\theta_{v, w})\|v\|\|w\|  $$
- Hai công thức đưa ra về mặt đại số và hình học là tương đương với nhau

```python
fig, ax = plt.subplots(nrows = 1, ncols= 5)
fig.set_figheight(5)
fig.set_figwidth(20)

angle = [30, 120, 90, 0, 180]

for i in range(len(angle)) : 
    x_1 , y_1 = create_vector(2, 0)
    x_2, y_2 = create_vector(2, angle[i])
    ax[i].quiver(0, 0, x_1, y_1, angles='xy', scale_units='xy', scale=1, color='blue')
    ax[i].quiver(0, 0, x_2, y_2, angles = 'xy', scale_units='xy', scale=1, color='green')
    ax[i].set_aspect(True)
    ax[i].set_xlim([-3, 3])
    ax[i].set_ylim([-3,  3])
    ax[i].grid(True)
    ax[i].set_title(f'alpha = {angle[i]} \n dot product = {round(np.cos(np.radians(angle[i])) * 2 * 2)}')
plt.show()
```

## 1. Correlation and cosine similarity
- Correlation coefficient là một số dùng để đo tính tuyến tính giữa 2 biến, nó có giới hạn trong khoảng từ -1 đến 1 với -1 là ngược chiều, 1 là cùng chiều và 0 là không có tính tuyến tính

![Local Image](Img/Figure4_1.png) 

- Có hai cách chuẩn hóa : 
    - Mỗi giá trị dữ liệu, trừ đi giá trị trung bình của nó 
    - Chia dot product bởi tích của các vector chuẩn hóa.
- Công thức tính Pearson correlation coefficient : $$ p = \frac {\sum ^n _{i = 1} (x_i - \bar x)(y_i - \bar y)} {\sqrt{\sum ^n _{i = 1} (x_i - \bar x)^2} \sqrt {\sum ^n _{i = 1}(y_i - \bar y)  ^2}} $$

- Công thức biểu diễn thông qua dot product : $$ p = \frac {\tilde {x^T} \tilde y} {\| \tilde x \| \| \tilde y \|} $$ Trong đó $\tilde x$ là hiệu mỗi giá trị với giá trị  trung bình 

- Công thức cosine similarity là công thức hình học của dot product ($\alpha$ là dot product của v và w) :  $$ cos(\theta _{v, w}) = \frac {\alpha} {\| v\| \| w \| } $$

```python
v = np.array([0, 1, 2, 3])
w = np.array([100, 101, 102, 103])

cosine = np.dot(v, w) / (np.linalg.norm(v) * np.linalg.norm(w))
print(cosine)
```

## k-Means clustering
- k-means clustering là phương pháp không giám sát phân loại dữ liệu vào một số nhóm liên quan dựa trên khoảng cách ngắn nhất giữa chúng.
- Thuật toán k-means đơn giản : 
    1. Tạo ra k centroids như một điểm ngẫu nhiên trong không gian dữ liệu (centroid là lớp,  nhóm cần được đưa vào phân loại)
    2. Tính toán khoảng cách euclid từ điểm quan sát tới centroid 
    3. Gán điểm quan sát vào nhóm có khoảng cách nhỏ nhất 
    4. Cập nhật trung bình dữ liệu quan sát được gán vào mỗi centroid 

```python
data = np.random.randn(150, 2)

# Step 1 : Create k centroids as randoms points in the data space
k = 3 
ridx = np.random.choice(range(len(data)), k, replace = False)
centroids = data[ridx, :]

centroids
```

```python
# Step 2: Compute the distance between data observation to centroid 
dists = np.zeros((data.shape[0], k))

for ci in range(k) : 
    dists[:, ci] = np.sum((data - centroids[ci, :]) ** 2, axis = 1)
dists
```

```python
# Step 3 : to assign each data observation to the group with minimum distance
groupidx = np.argmin(dists, axis = 1)
groupidx
```

```python
# Step 4: to recompute the centroids as the mean of all data points within the class
for ki in range(k) : 
    centroids[ki, :] = [np.mean(data[groupidx == ki, 0]), 
                        np.mean(data[groupidx == ki, 1])]
centroids
```

```python
fig, ax = plt.subplots(figsize = (20, 10))

ax.scatter(data[:,0], data[:,1])
ax.scatter(centroids[:, 0], centroids[:, 1], color = 'red')
for i in range(len(groupidx)) : 
    ax.plot([centroids[groupidx[i], 0], data[i, 0]], [centroids[groupidx[i], 1], data[i, 1]], color = 'green')

plt.show()
```
